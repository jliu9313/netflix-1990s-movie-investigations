```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")

# Filter for movies from the 1990s
movies_1990s = netflix_df[
    (netflix_df["type"] == "Movie") &
    (netflix_df["release_year"] >= 1990) &
    (netflix_df["release_year"] < 2000)
]

# Most frequent movie duration in the 1990s
duration = int(movies_1990s["duration"].mode()[0])

# Count short action movies (< 90 min) from the 1990s
short_movie_count = int(movies_1990s[
    (movies_1990s["genre"] == "Action") &
    (movies_1990s["duration"] < 90)
].shape[0])

print(duration)
print(short_movie_count)

# Top 5 most common countries
top_5_countries = movies_1990s["country"].value_counts().head(5)
print(top_5_countries)

from collections import Counter

# Combine all titles into one string, split into words, and lowercase
words = " ".join(netflix_df["title"].dropna()).lower().split()

# Remove common stop words (optional but recommended)
stop_words = {"the", "a", "an", "of", "and", "in", "to", "is", "on", "de", "2", "my", "&", "i", "for", "you"}
filtered_words = [word for word in words if word not in stop_words]

# Count and get top 5
top_5_words = Counter(filtered_words).most_common(5)
print(top_5_words)

# Top 3 most common genres
top_3_genres = movies_1990s["genre"].value_counts().head(3)
print(top_3_genres)

# Top 5 most common directors
top_5_directors = movies_1990s["director"].value_counts().head(5)
print(top_5_directors)

import matplotlib.pyplot as plt

# Get top 3 genres
top_3_genres = netflix_df["genre"].value_counts().head(3)

# Plot
plt.figure(figsize=(8, 5))
plt.bar(top_3_genres.index, top_3_genres.values, color=["steelblue", "coral", "mediumseagreen"])
plt.title("Top 3 Most Common Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

import matplotlib.pyplot as plt

# Get top 5 countries
top_5_countries = netflix_df["country"].value_counts().head(5)

# Plot
plt.figure(figsize=(8, 5))
plt.bar(top_5_countries.index, top_5_countries.values, color=["steelblue", "coral", "mediumseagreen", "mediumpurple", "goldenrod"])
plt.title("Top 5 Countries of Origin on Netflix")
plt.xlabel("Country")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
